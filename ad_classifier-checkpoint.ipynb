## Problem Statement

The objective of this project is to develop a machine learning-based classification system that can detect Alzheimer’s disease from EEG signal features. By leveraging algorithms such as XGBoost, the system aims to accurately distinguish between healthy individuals and Alzheimer’s patients, enabling a potential tool for early and scalable screening.

## 5. Train–Test Split (Leakage Protection)

```python
frontal = ['Fp1','Fp2','F7','F3','Fz','F4','F8']
central = ['C3','Cz','C4']
temporal = ['T3','T4','T5']
parietal = ['P3','Pz','P4']

df['frontal_mean'] = df[frontal].mean(axis=1)
df['central_mean'] = df[central].mean(axis=1)
df['temporal_mean'] = df[temporal].mean(axis=1)
df['parietal_mean'] = df[parietal].mean(axis=1)
```

```python
X = df.drop(columns=['status'])
y = df['status']

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)
print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)
```

## 6. Outlier Clipping (1% Extreme Values – Train Only)

```python
# X_train_clipped = X_train.copy()
# X_test_clipped = X_test.copy()

# train_rows_clipped = set()
# test_rows_clipped = set()

# for col in eeg_channels:
#     lower = X_train_clipped[col].quantile(0.01)
#     upper = X_train_clipped[col].quantile(0.99)

#     # Identify affected rows BEFORE clipping
#     train_mask = (X_train_clipped[col] < lower) | (X_train_clipped[col] > upper)
#     test_mask  = (X_test_clipped[col] < lower) | (X_test_clipped[col] > upper)

#     train_rows_clipped.update(X_train_clipped.index[train_mask])
#     test_rows_clipped.update(X_test_clipped.index[test_mask])

#     # Apply clipping
#     X_train_clipped[col] = X_train_clipped[col].clip(lower, upper)
#     X_test_clipped[col]  = X_test_clipped[col].clip(lower, upper)

#     print(f"{col}: clipped to range [{lower:.2f}, {upper:.2f}]")

# print("\nClipping completed successfully.")
# print(f"Total TRAIN rows affected: {len(train_rows_clipped)} "
#       f"({len(train_rows_clipped)/len(X_train_clipped)*100:.2f}%)")
# print(f"Total TEST rows affected: {len(test_rows_clipped)} "
#       f"({len(test_rows_clipped)/len(X_test_clipped)*100:.2f}%)")
X_train_clipped = X_train.copy()
X_test_clipped = X_test.copy()

for col in eeg_channels:
    lower = X_train_clipped[col].quantile(0.01)
    upper = X_train_clipped[col].quantile(0.99)

    X_train_clipped[col] = X_train_clipped[col].clip(lower, upper)
    X_test_clipped[col] = X_test_clipped[col].clip(lower, upper)
```

```python
scaler = MinMaxScaler()

X_train_scaled = scaler.fit_transform(X_train_clipped)
X_test_scaled = scaler.transform(X_test_clipped)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
import lightgbm as lgb
from sklearn.metrics import accuracy_score

models = {
    # "LightGBM": lgb.LGBMClassifier(
    #     n_estimators=80, max_depth=3, num_leaves=7, learning_rate=0.05,
    #     min_child_samples=100, subsample=0.6, colsample_bytree=0.6,
    #     class_weight="balanced", random_state=42
    # ),
    "XGBoost": XGBClassifier(
    n_estimators=60,        # between 50 and 80
    max_depth=3,            # keep at 2
    learning_rate=0.04,     # between 0.03 and 0.05
    subsample=0.6,          # back to original
    colsample_bytree=0.6,   # back to original
    tree_method='hist',
    eval_metric='logloss',
    random_state=RANDOM_SEED
),
    "Random Forest": RandomForestClassifier(
        n_estimators=100, max_depth=5, n_jobs=-1,
        class_weight="balanced", random_state=42
    ),
    "Logistic Regression": LogisticRegression(
        max_iter=1000, class_weight="balanced", n_jobs=-1, random_state=42
    )
}

# for name, model in models.items():
#     model.fit(X_train_scaled, y_train)
    
#     train_preds = model.predict(X_train_scaled)
#     test_preds  = model.predict(X_test_scaled)
    
#     train_acc = accuracy_score(y_train, train_preds)
#     test_acc  = accuracy_score(y_test, test_preds)
    
#     print(f"{name}")
#     print(f"  Train Accuracy: {train_acc:.4f}")
#     print(f"  Test Accuracy:  {test_acc:.4f}")
#     print()
results = []

for name, model in models.items():
    model.fit(X_train_scaled, y_train)

    train_preds = model.predict(X_train_scaled)
    test_preds  = model.predict(X_test_scaled)

    train_acc = accuracy_score(y_train, train_preds)
    test_acc  = accuracy_score(y_test, test_preds)

    results.append({
        "Model": name,
        "Train Accuracy": train_acc,
        "Test Accuracy": test_acc
    })

    print(f"{name}")
    print(f"  Train Accuracy: {train_acc:.4f}")
    print(f"  Test Accuracy:  {test_acc:.4f}")
    print()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

results_df = pd.DataFrame(results)

results_df.set_index("Model")[["Train Accuracy","Test Accuracy"]].plot(
    kind="bar",
    figsize=(8,5)
)

plt.title("Model Accuracy Comparison")
plt.ylabel("Accuracy")
plt.xticks(rotation=0)
plt.ylim(0.8,1.0)

plt.show()
```
